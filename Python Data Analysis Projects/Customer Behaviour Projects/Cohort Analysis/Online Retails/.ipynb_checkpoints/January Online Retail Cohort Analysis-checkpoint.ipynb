```python
# Lets Do Preprocessing and cleaning

# convert the data tyepes
retail['Invoice'] = retail['Invoice'].astype(int)
retail['Price '] = retail['Price '].astype(int)
retail['Customer ID '] = retail['Customer ID'].astype(int)
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'])
retail = retail.sort_values('InvoiceDate')

# lets create total column
retail['Total'] = retail['Quantity'] * retail['Price']
```
