```python
# Let import librries

import pandas as pd
import numpy as np

# load the data
espn_cricket_df = pd.read_csv(r"C:\Users\jki\Downloads\ESP Cricket Data.csv")
espn_cricket_df.head(10)
```

```python
# let check for missing values
missing_values = espn_cricket_df.isna().sum()
print(missing_values)
```

```python
# lets have look at the data types
espn_cricket_df.info()
```

```python
# rename columns
espn_cricket_df =espn_cricket_df.rename(columns = {'Span':'Career Period','Matces':'Matches','HS':'Highest_Inns_Score','BF':'Balls_Faced','SR':'Batting_Strke_Rate'})
espn_cricket_df.head(5)
```

```python
# Letch check for null values
espn_cricket_df.head(5).isnull().any()
```

```python
# let check for duplicate
espn_cricket_df.duplicated()
```

```python
espn_cricket_df.head(4)
```

```python
# let check on the data type of Career Perios

# Assuming espn_cricket_df is your DataFrame


# Split the 'Career Period' column into 'Start Year' and 'End Year'
espn_cricket_df[['Rookie Year', 'Final Year']] = espn_cricket_df['Career Period'].str.split('-', 1, expand=True)

# Display the updated DataFrame

espn_cricket_df.head(5)
```

```python
# lets convert the rookie year and final years into intergers
espn_cricket_df['Rookie Year'] =pd.to_numeric(espn_cricket_df['Rookie Year'])
espn_cricket_df['Final Year'] =pd.to_numeric(espn_cricket_df['Final Year'])

# lets check on data types
espn_cricket_df['Rookie Year'].info()
espn_cricket_df['Final Year'].info()
```

```python
# let create the years played columns
espn_cricket_df['Years Played'] = espn_cricket_df['Final Year'] - espn_cricket_df['Rookie Year']
espn_cricket_df.head(5)
```

```python
# lets get the data types of the data 
espn_cricket_df.info()
```

```python
# let check on the statistsal summary
espn_cricket_df.head()
```

```python
# Lets change the data type of Inner Score and calculate the avaerage strike rate
# lets clean up strike rate


# Removing asterisks from the 'average' column
espn_cricket_df['Highest_Inns_Score'] = espn_cricket_df['Highest_Inns_Score'].str.replace('*', '')

# Converting the 'average' column to numeric type
espn_cricket_df['Highest_Inns_Score'] = pd.to_numeric(espn_cricket_df['Highest_Inns_Score'])

# Displaying the result
print(espn_cricket_df['Highest_Inns_Score'])
```

## 3 .Whats the  Average_Strike Rate for cricketers who played over 10 years


```python
espn_cricket_df['Batting_Strike_Rate'] =espn_cricket_df['Batting_Strke_Rate']  


# Assuming espn_cricket_df is your DataFrame
# Replace "-" with NaN in the 'Batting_Strike_Rate' column
espn_cricket_df['Batting_Strike_Rate'] = espn_cricket_df['Batting_Strike_Rate'].replace('-', pd.NA, regex=True)

# If you want to remove rows with "-" in the 'Batting_Strike_Rate' column
# espn_cricket_df = espn_cricket_df[espn_cricket_df['Batting_Strike_Rate'] != '-']

# If you want to convert the column to numeric after replacing "-"
espn_cricket_df['Batting_Strike_Rate'] = pd.to_numeric(espn_cricket_df['Batting_Strike_Rate'], errors='coerce')

# lets get te the average strike rate above 10 years
espn_cricket_df[espn_cricket_df['Years Played']>10 ]['Batting_Strike_Rate'].mean()
```

The Average Strike Rate is 63

## 5  The Max Highest Inns Score by country

```python
espn_cricket_df.head(5)



espn_cricket_df.groupby("Country")['Highest_Inns_Score'].max().to_frame("High_in_score_by_Country").reset_index().sort_values("High_in_score_by_Country")

```

## 5  The Average  Matches  by country

```python
espn_cricket_df.groupby("Country")['Matches'].mean().to_frame("Average_in_score_by_Country").reset_index().sort_values("Average_in_score_by_Country")

```
