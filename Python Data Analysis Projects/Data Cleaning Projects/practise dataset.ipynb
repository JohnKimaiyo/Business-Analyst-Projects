```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Year_of_Release': [2006, 1985, 2008, 2009, 1996, 1989, 2006, 2006, 2009, 1984, 2005, 2005, 1999],
        'NA_sales': [41.36, 29.08, 15.68, 15.61, 11.27, 23.2, 11.28, 13.96, 14.44, 26.93, 9.05, 9.71, 9]}

# Create a DataFrame
df = pd.DataFrame(data)

# Group by 'Year_of_Release' and calculate total 'NA_sales' for each year
yearly_sales = df.groupby('Year_of_Release')['NA_sales'].sum().reset_index()

# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(yearly_sales['Year_of_Release'], yearly_sales['NA_sales'], color='skyblue')
plt.xlabel('Year of Release')
plt.ylabel('NA Sales (in millions)')
plt.title('NA Sales for Each Year of Release')
plt.show()
```
