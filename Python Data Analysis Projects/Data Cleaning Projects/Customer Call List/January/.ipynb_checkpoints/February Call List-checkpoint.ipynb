```python
# lets import libraries
import pandas as pd
import numpy as np

call_list = pd.read_excel(r"C:\Users\jki\Downloads\Customer Call List.xlsx")
call_list
```

```python
call_list = call_list.drop_duplicates()
call_list
```

```python
# lets drop not useful column
call_list = call_list.drop(columns = "Not_Useful_Column")
call_list
```

```python
# lets remove  strings
call_list["Last_Name"] = call_list["Last_Name"].str.strip("123._/")
call_list
```

```python
# let drop null values
call_list["Phone_Number"] = call_list["Phone_Number"].str.replace('nan--','')
call_list["Phone_Number"] = call_list["Phone_Number"].str.replace('Na--','')
call_list
```

```python
# lets sepaate the address column
call_list[["Street_Address","State","Zip_code"]] =  call_list["Address"].str.split(',',2, expand = True)
call_list
```

```python
# change the column 
call_list["Do_Not_Contact"] = call_list["Do_Not_Contact"].str.replace('Yes','Y')
call_list["Do_Not_Contact"] = call_list["Do_Not_Contact"].str.replace('No','Y')
call_list


```

```python
# Lets replace N/A with empty stricng
call_list = call_list.fillna("")
call_list
```

```python
# in the do not conatct column replace the empty string to Y

for x in call_list.index:
    if call_list.loc[x,"Do_Not_Contact"] == '':
        call_list.drop(x,inplace =True)
call_list
```

```python
# lets drop null values in the phone contacts
for x in call_list.index:
    if call_list[x,"Phone_Number"] == '':
        call_list.drop(x, inplace =True)
call_list
```

```python
#
call_list = call_list.reset_index(drop=True)
call_list
```
