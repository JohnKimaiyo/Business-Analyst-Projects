```python
# import libraries
import pandas as pd
import numpy as np


# lets load the dataset
call_list = pd.read_excel(r"C:\Users\jki\Downloads\Customer Call List.xlsx")
call_list
```

```python
#lets check for missing values
missing_values = call_list.isna().sum()
print(missing_values)
```

```python
# lets remove the missing values
call_list.dropna(subset=["Last_Name"],inplace=True)
call_list.dropna(subset=["Phone_Number"],inplace=True)
call_list.dropna(subset=["Do_Not_Contact"],inplace=True)

#lets check for missing values
missing_values = call_list.isna().sum()
print(missing_values)

```

```python
# lets drop duplicates
call_list=call_list.drop_duplicates()
call_list
```

```python
# lets drop column not need
call_list= call_list.drop(columns="Not_Useful_Column")
call_list
```
