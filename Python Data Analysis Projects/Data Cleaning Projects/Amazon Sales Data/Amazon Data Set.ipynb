```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None


# Now we need to read in the data
df = pd.read_csv(r'C:\Users\jki\Desktop\Python Customer Sgementation Project Zero\Python Data Analysis Projects\Data Cleaning Projects\Amazon Sales Data\Data set\marketing_sample_for_amazon_com-ecommere.csv')
```

```python
# Now let's take a look at the data

df
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Are there any Outliers?

df.boxplot(column=['Selling Price'])
```

```python
df.drop_duplicates()
```

```python
# Order our Data a little bit to see

df.sort_values(by=['Selling Price'], inplace=False, ascending=False)
```

```python
sns.regplot(x="Selling Price", y="List Price", data=df)
```

```python
# Correlation Matrix between all numeric columns

df.corr(method ='pearson')
```

```python
df.corr(method ='kendall')
```

```python
df.corr(method ='spearman')
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Amazon Sales")

plt.xlabel("Selling Price")

plt.ylabel("Brand Name")

plt.show()
```
