```python

# Load in some packages
import calendar
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# load dataset
bigmart_sales_df = pd.read_csv(r"C:\Users\jki\Downloads\2  BigMart\Source Files\Cleaned Data\Big Mart Sales.csv" )  
bigmart_sales_df
```

```python
# lets check for missing values
missing_values = bigmart_sales_df.isna().sum()
print(missing_values)
```

```python
# lets remove the missing values

bigmart_sales_df.dropna(subset=['Weight'],inplace=True)
bigmart_sales_df.dropna(subset=['Quantity'],inplace=True)
bigmart_sales_df.dropna(subset=['OutletSize'],inplace=True)
# lets check for missing values
missing_values = bigmart_sales_df.isna().sum()
print(missing_values)
```

```python
bigmart_sales_df.head(6)
```

```python
# lets see the datatypes
bigmart_sales_df.info()
```

```python
# lets create sales amaount

bigmart_sales_df['Unit Price'] = bigmart_sales_df['MRP']
bigmart_sales_df.describe()

```

```python
# lets create Sales Amout Colummn

bigmart_sales_df['Sales Amount'] =  bigmart_sales_df['Quantity'] * bigmart_sales_df['Unit Price']
bigmart_sales_df.describe()
```

```python
# let have a look at the data types
bigmart_sales_df.info()
```

```python
# lets extract the month from date
bigmart_sales_df['Date'] = pd.to_datetime(bigmart_sales_df['Date'])
bigmart_sales_df.info()
```

```python
# lets create month colum from date column
bigmart_sales_df['Year'] = bigmart_sales_df['Date'].dt.year
bigmart_sales_df.head()
```

```python

  # Load in some packages
import calendar
import warnings
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations
from collections import Counter

warnings.filterwarnings("ignore")

# 1. What was the best month for sales? How much was earned that Year?

# Replace NaN or inf values in the 'Month' column with a default value (e.g., 0)
bigmart_sales_df['Year'] = bigmart_sales_df['Year'].fillna(0).astype(int)

# Group by Year and calculate total sales for each Year
sales_by_year = bigmart_sales_df.groupby('Year').sum()['Sales Amount']

# Find the best month for sales and the corresponding earnings
best_year = sales_by_year.idxmax()
earnings_for_best_year = sales_by_year.max()

print(f"The best Year for sales was {best_year} with earnings of ${earnings_for_best_year:,.2f}")
```
