```python

# Load in some packages
import calendar
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

# lets load the dataset
bigmart_sales_df = pd.read_excel(r"C:\Users\jki\Downloads\2  BigMart\Source Files\Cleaned Data\Big Mart Sales.xlsx")
bigmart_sales_df
```

```python
# lets remove the missing values
missing_values = bigmart_sales_df.isna().sum()
print(missing_values)
```

```python
# lets drop the missing values
bigmart_sales_df.dropna(subset=['Quantity'],inplace=True)
bigmart_sales_df.dropna(subset=['Weight'],inplace=True)
bigmart_sales_df.dropna(subset=['OutletSize'],inplace=True)

# lets remove the missing values
missing_values = bigmart_sales_df.isna().sum()
print(missing_values)
```

```python
# lets convert the date in date type
bigmart_sales_df['Date'] =  pd.to_datetime(bigmart_sales_df['Date'])
bigmart_sales_df['Date'].info()
```
