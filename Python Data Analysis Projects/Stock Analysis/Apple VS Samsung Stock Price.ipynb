## Here's a Python script that fetches historical stock prices for Delta and American Airlines using the yfinance library and compares them:

```python

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock symbols for Delta Air Lines and American Airlines
delta_symbol = "DAL"
american_symbol = "AAL"

# Define the time period for stock price data (e.g., last 5 years)
start_date = "2019-01-01"
end_date = "2023-01-01"

# Fetch stock price data using yfinance
delta_data = yf.download(delta_symbol, start=start_date, end=end_date)
american_data = yf.download(american_symbol, start=start_date, end=end_date)

# Extract the 'Close' prices
delta_close_prices = delta_data['Close']
american_close_prices = american_data['Close']

# Create a DataFrame to store the stock prices
stock_prices_df = pd.DataFrame({
    'Delta': delta_close_prices,
    'American': american_close_prices
})

# Plot the stock prices
stock_prices_df.plot(title='Delta vs American Airlines Stock Prices', ylabel='Stock Price (USD)', xlabel='Date')
plt.show()

# Calculate and display summary statistics
summary_statistics = stock_prices_df.describe()
print(summary_statistics)

# Calculate and display the correlation between Delta and American Airlines stock prices
correlation = stock_prices_df['Delta'].corr(stock_prices_df['American'])
print(f"Correlation between Delta and American Airlines stock prices: {correlation}")

# Calculate and display the percentage change in stock prices
percentage_change = stock_prices_df.pct_change()
print(percentage_change.head())

```
