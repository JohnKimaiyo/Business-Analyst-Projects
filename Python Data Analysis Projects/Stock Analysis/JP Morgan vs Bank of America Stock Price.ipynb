```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock symbols for JP Morgan and Bank of America
jpm_symbol = "JPM"
boa_symbol = "BAC"

# Define the time period for stock price data (e.g., last 5 years)
start_date = "2019-01-01"
end_date = "2023-01-01"

# Fetch stock price data using yfinance
jpm_data = yf.download(jpm_symbol, start=start_date, end=end_date)
boa_data = yf.download(boa_symbol, start=start_date, end=end_date)

# Extract the 'Close' prices
jpm_close_prices = jpm_data['Close']
boa_close_prices = boa_data['Close']

# Create a DataFrame to store the stock prices
stock_prices_df = pd.DataFrame({
    'JP Morgan': jpm_close_prices,
    'Bank of America': boa_close_prices
})

# Plot the stock prices
stock_prices_df.plot(title='JP Morgan vs Bank of America Stock Prices', ylabel='Stock Price (USD)', xlabel='Date')
plt.show()

# Calculate and display summary statistics
summary_statistics = stock_prices_df.describe()
print(summary_statistics)

# Calculate and display the correlation between JP Morgan and Bank of America stock prices
correlation = stock_prices_df['JP Morgan'].corr(stock_prices_df['Bank of America'])
print(f"Correlation between JP Morgan and Bank of America stock prices: {correlation}")

# Calculate and display the percentage change in stock prices
percentage_change = stock_prices_df.pct_change()
print(percentage_change.head())
```
