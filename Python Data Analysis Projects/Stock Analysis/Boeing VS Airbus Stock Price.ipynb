## To show the relationship between  Boeing and Airbus, you can use historical data for both and plot them on the same graph. Here's an example Python script using Pandas and Matplotlib:

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock symbols for Boeing and Airbus
boeing_symbol = "BA"
airbus_symbol = "AIR.PA"  # Airbus stock symbol on Yahoo Finance

# Define the time period for stock price data (e.g., last 5 years)
start_date = "2019-01-01"
end_date = "2023-01-01"

# Fetch stock price data using yfinance
boeing_data = yf.download(boeing_symbol, start=start_date, end=end_date)
airbus_data = yf.download(airbus_symbol, start=start_date, end=end_date)

# Extract the 'Close' prices
boeing_close_prices = boeing_data['Close']
airbus_close_prices = airbus_data['Close']

# Create a DataFrame to store the stock prices
stock_prices_df = pd.DataFrame({
    'Boeing': boeing_close_prices,
    'Airbus': airbus_close_prices
})

# Plot the stock prices
stock_prices_df.plot(title='Boeing vs Airbus Stock Prices', ylabel='Stock Price (USD)', xlabel='Date')
plt.show()

# Calculate and display summary statistics
summary_statistics = stock_prices_df.describe()
print(summary_statistics)

# Calculate and display the correlation between Boeing and Airbus stock prices
correlation = stock_prices_df['Boeing'].corr(stock_prices_df['Airbus'])
print(f"Correlation between Boeing and Airbus stock prices: {correlation}")

# Calculate and display the percentage change in stock prices
percentage_change = stock_prices_df.pct_change()
print(percentage_change.head())

```
