```python

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock symbols for Amazon and Walmart
amazon_symbol = "AMZN"
walmart_symbol = "WMT"

# Define the time period for stock price data (e.g., last 5 years)
start_date = "2019-01-01"
end_date = "2023-01-01"

# Fetch stock price data using yfinance
amazon_data = yf.download(amazon_symbol, start=start_date, end=end_date)
walmart_data = yf.download(walmart_symbol, start=start_date, end=end_date)

# Extract the 'Close' prices
amazon_close_prices = amazon_data['Close']
walmart_close_prices = walmart_data['Close']

# Create a DataFrame to store the stock prices
stock_prices_df = pd.DataFrame({
    'Amazon': amazon_close_prices,
    'Walmart': walmart_close_prices
})

# Plot the stock prices
stock_prices_df.plot(title='Amazon vs Walmart Stock Prices', ylabel='Stock Price (USD)', xlabel='Date')
plt.show()

# Calculate and display summary statistics
summa
```
