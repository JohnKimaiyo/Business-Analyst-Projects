```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock symbols for Eli Lilly and Company and Johnson & Johnson
lly_symbol = "LLY"
jnj_symbol = "JNJ"

# Define the time period for stock price data (e.g., last 5 years)
start_date = "2019-01-01"
end_date = "2023-01-01"

# Fetch stock price data using yfinance
lly_data = yf.download(lly_symbol, start=start_date, end=end_date)
jnj_data = yf.download(jnj_symbol, start=start_date, end=end_date)

# Extract the 'Close' prices
lly_close_prices = lly_data['Close']
jnj_close_prices = jnj_data['Close']

# Create a DataFrame to store the stock prices
stock_prices_df = pd.DataFrame({
    'Eli Lilly and Company': lly_close_prices,
    'Johnson & Johnson': jnj_close_prices
})

# Plot the stock prices
stock_prices_df.plot(title='Eli Lilly and Company vs Johnson & Johnson Stock Prices', ylabel='Stock Price (USD)', xlabel='Date')
plt.show()

# Calculate and display summary statistics
summary_statistics = stock_prices_df.describe()
print(summary_statistics)

# Calculate and display the correlation between Eli Lilly and Company and Johnson & Johnson stock prices
correlation = stock_prices_df['Eli Lilly and Company'].corr(stock_prices_df['Johnson & Johnson'])
print(f"Correlation between Eli Lilly and Company and Johnson & Johnson stock prices: {correlation}")

# Calculate and display the percentage change in stock prices
percentage_change = stock_prices_df.pct_change()
print(percentage_change.head())


```
