```python
import pandas as pd

# Create a DataFrame with a column of strings containing extra spaces
data = {'Name': ['   John', 'Alice  ', '  Bob   ']}
df = pd.DataFrame(data)

# Use df['Name'].str.strip() to remove leading and trailing spaces
df['Name'] = df['Name'].str.strip()

# Display the updated DataFrame
print(df)
```
